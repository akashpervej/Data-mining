```python
import matplotlib.pyplot as plt
plt.hist(df['ApplicantIncome'],bins=5, rwidth=0.8)
```

**Histrogram will show the frequency of 'Applicant Income' values within each bin. This can help us see how the data is distributed, whether it is skewed or symmetric, and whether there are any outliers or unusual patterns.**

```python
plt.hist(df['LoanAmount'],bins=5, rwidth=0.8)
```

**Histogram can be useful for visualizing the distribution of 'LoanAmount' values in the dataset. It can show, for example, if the data is skewed towards higher or lower values, if there are any outliers, or if the data is normally distributed. By adjusting the number of bins or the width of the bars, the visualization can be tailored to highlight specific aspects of the data.**

```python
sns.distplot(df['ApplicantIncome'], bins=5, kde=True)
```

**Used to quickly visualize the distribution of values in the 'ApplicantIncome' column of a DataFrame df using both a histogram and a KDE plot. This can help to identify the overall shape of the distribution and to gain a sense of the central tendency and variability of the data.**

```python
sns.boxplot(df['LoanAmount'])
```

**In this specific case, the boxplot will show the distribution of the loan amounts in the 'LoanAmount' column of the DataFrame.**

```python
sns.violinplot(x='LoanAmount',data=df)
```

**A violin plot is a type of plot that combines elements of a box plot and a kernel density plot. It shows the distribution of the data as a density curve, with the shape of the curve indicating the frequency of observations at different values. The thicker parts of the curve represent areas where more observations occur, while the thinner parts represent areas where fewer observations occur.**

```python
sns.scatterplot(data=df, x="ApplicantIncome", y="LoanAmount")
```

**A scatter plot is a type of data visualization that uses dots to represent the data points on a two-dimensional plane. It is useful for exploring the relationship between two numerical variables, as it allows us to see the distribution of points and any patterns that may exist.**

```python
sns.scatterplot(data=df, x="ApplicantIncome", y="LoanAmount",hue='Gender',style='Self_Employed', size = "Property_Area")
```

**The plot will have ApplicantIncome on the x-axis and LoanAmount on the y-axis, with different data points represented by different colors based on the 'Gender' variable. Additionally, the data points will be styled based on the 'Self_Employed' variable, where different marker styles will be used for self-employed individuals and non-self-employed individuals. Finally, the size of the data points will be determined by the 'Property_Area' variable**

```python
sns.barplot(data=df,x='Property_Area',y='ApplicantIncome')
```

**The resulting bar plot can help identify any patterns or trends in the relationship between the Property_Area and ApplicantIncome variables. For example, it can help answer questions like: Are applicants from certain property areas more likely to have higher incomes? Are there any significant differences in the incomes of applicants from different property areas?**

```python
sns.barplot(data=df,x='Property_Area',y='ApplicantIncome',hue='Gender')
```

**That shows the average "ApplicantIncome" for each combination of "Property_Area" and "Gender" values in the "df" dataframe. The plot will use different colors to distinguish between male and female applicants. This can help identify any patterns or trends in the data and visually compare the average "ApplicantIncome" between different groups.**

```python
sns.boxplot(data=df,x='Gender',y='LoanAmount')
```

**The resulting visualization will show a box for each category of gender (e.g. male and female) and the distribution of loan amounts for each gender. The box represents the middle 50% of the data, with the bottom and top edges representing the first and third quartiles (Q1 and Q3). The line within the box represents the median, and the whiskers extending from the box indicate the range of the data outside of the middle 50%.**

```python
sns.boxplot(data=df,x='Gender',y='LoanAmount', hue= "Self_Employed")
```

**The resulting plot will show box-and-whisker plots for each combination of gender and self-employment status. The boxes represent the interquartile range (IQR) of the loan amount distribution for each group, while the whiskers represent the range of values within 1.5 times the IQR. Any outliers beyond this range will be plotted as individual points. The hue parameter allows us to distinguish between self-employed and non-self-employed individuals within each gender group.**

```python
sns.distplot(df[df['Property_Area']=='Urban']['LoanAmount'],hist=False,color="red")
sns.distplot(df[df['Property_Area']=='Rural']['LoanAmount'],hist=False,color="black")
sns.distplot(df[df['Property_Area']=='Semiurban']['LoanAmount'],hist=False)
```

**The first plot is the distribution of LoanAmount for properties located in Urban areas, shown in red. The second plot shows the distribution of LoanAmount for properties located in Rural areas, shown in black. The third plot shows the distribution of LoanAmount for properties located in Semiurban areas, with the default color blue.**

```python
sns.heatmap(pd.crosstab(df['Property_Area'],df['Self_Employed']))
```

**Heatmaps are useful for visualizing the relationships between two categorical variables, especially when there are many unique values for each variable. They can help identify patterns and trends in the data, and can be useful for exploratory data analysis and hypothesis testing. In this case, the heatmap could help to identify any potential association between 'Property_Area' and 'Self_Employed', by highlighting which combinations of values are more or less common.**

```python
sns.pairplot(df)
```

**By visualizing all pairs of variables in a dataset, pairplot() provides a quick and convenient way to identify patterns and relationships between variables. Each scatterplot shows the relationship between two variables by plotting their values against each other. The histograms show the distribution of each variable individually.**

```python
sns.lineplot(data=df, x="LoanAmount", y="ApplicantIncome")
```

**Line plot is useful for visualizing the relationship between two variables, especially when one variable is continuous (like LoanAmount) and the other is numerical (like ApplicantIncome).
